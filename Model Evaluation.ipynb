# Cosine Similarity and BERT Score

```python
#pip install pandas bert-score scikit-learn
```

```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from bert_score import score


# Calculate BERTScore and cosine similarity for each row
bert_scores = []
cosine_similarities = []

for index, row in eval_df.iterrows():
    manual_text = row['Manual_ExtractedInfo']
    gpt_text = row['Claude_ExtractedInfo']
    
    # Calculate BERTScore
    _, _, bert_score = score([manual_text], [gpt_text], lang='en')
    bert_score = bert_score.item()  # Extracting the actual score from tensor
    bert_scores.append(bert_score)
    
    # Calculate cosine similarity
    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform([manual_text, gpt_text])
    cosine_sim = cosine_similarity(tfidf_matrix)[0][1]
    cosine_similarities.append(cosine_sim)

# Add scores to the DataFrame
eval_df['Claude_BERTScore'] = bert_scores
eval_df['Claude_CosineSimilarity'] = cosine_similarities


```
